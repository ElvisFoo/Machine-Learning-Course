We split the dataset now for training and validation purposes

```python
from sklearn.model_selection import train_test_split
```

```python
train_val_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
train_df, val_df = train_test_split(train_val_df, test_size=0.25, random_state=42)
```

```python
print("train_df.shape:", train_df.shape)
print("val_df.shape:", val_df.shape)
print("test_df.shape:", test_df.shape)
```

But since this is time series data, we can train on older data, validate and test on the later data. Hence ignoring what we did above:

```python
year = pd.to_datetime(df['Date']).dt.year

train_df = df[year < 2015]
val_df = df[year == 2015]
test_df = df[year > 2015]
```

```python
print("train_df.shape:", train_df.shape)
print("val_df.shape:", val_df.shape)
print("test_df.shape:", test_df.shape)
```

```python
input_cols = list(train_df.columns)[1:-1]
target_col = 'RainTomorrow'
print("Input columns:", input_cols)
print("Target column:", target_col)
```

```python
train_inputs = train_df[input_cols].copy()
train_targets = train_df[target_col].copy()

val_inputs = val_df[input_cols].copy()
val_targets = val_df[target_col].copy()

test_inputs = test_df[input_cols].copy()
test_targets = test_df[target_col].copy()
```

```python
numeric_cols = train_inputs.select_dtypes(include=np.number).columns.tolist()
categorical_cols = train_inputs.select_dtypes(include='object').columns.tolist()
print("Numeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)
```

```python
train_inputs[numeric_cols].describe()
```

```python
train_inputs[categorical_cols].nunique()
```

We replace missing values

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean')
```

```python
train_inputs[numeric_cols].isna().sum()
```

```python
imputer.fit(train_inputs[numeric_cols])
train_inputs[numeric_cols] = imputer.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = imputer.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = imputer.transform(test_inputs[numeric_cols])
```

```python
train_inputs[numeric_cols].isna().sum()
```

Replace missing values for categorical data as "Unknown"

```python
cat_imputer = SimpleImputer(strategy='constant', fill_value='Unknown')

cat_imputer.fit(train_inputs[categorical_cols])

train_inputs[categorical_cols] = cat_imputer.transform(train_inputs[categorical_cols])
val_inputs[categorical_cols] = cat_imputer.transform(val_inputs[categorical_cols])
test_inputs[categorical_cols] = cat_imputer.transform(test_inputs[categorical_cols])

train_inputs[categorical_cols].isna().sum()
```

We scale numeric values now so that no particular feature has a disproportionate impact on the model.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(train_inputs[numeric_cols])
train_inputs[numeric_cols] = scaler.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = scaler.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = scaler.transform(test_inputs[numeric_cols])
train_inputs[numeric_cols].describe()
```

Now we encode categorial data using one-hot encoding

```python
from sklearn.preprocessing import OneHotEncoder


encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoder.fit(train_inputs[categorical_cols])
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))
print(encoded_cols)
```

```python
train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical_cols])
val_inputs[encoded_cols] = encoder.transform(val_inputs[categorical_cols])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical_cols])

train_inputs = train_inputs[numeric_cols + encoded_cols]
val_inputs = val_inputs[numeric_cols + encoded_cols]
test_inputs = test_inputs[numeric_cols + encoded_cols]
```

```python
train_inputs.to_csv('train_inputs.csv', index=False)
train_targets.to_csv('train_targets.csv', index=False)
val_inputs.to_csv('val_inputs.csv', index=False)
val_targets.to_csv('val_targets.csv', index=False)
test_inputs.to_csv('test_inputs.csv', index=False)
test_targets.to_csv('test_targets.csv', index=False)
```

```python
train_inputs = pd.read_csv('train_inputs.csv')
train_targets = pd.read_csv('train_targets.csv')['RainTomorrow']
val_inputs = pd.read_csv('val_inputs.csv')
val_targets = pd.read_csv('val_targets.csv')['RainTomorrow']
test_inputs = pd.read_csv('test_inputs.csv')
test_targets = pd.read_csv('test_targets.csv')['RainTomorrow']
```

Training the logistic regression model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(train_inputs, train_targets)

print(model.coef_.tolist())
print(model.intercept_)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

def predict_and_plot(inputs, targets, name=''):
    preds = model.predict(inputs)
    accuracy = accuracy_score(targets, preds)
    print(f"{name} Accuracy: {accuracy}")
    cf=confusion_matrix(targets, preds, normalize='true')
    plt.figure()
    sns.heatmap(cf, annot=True, fmt='.2f')
    plt.xlabel('Prediction')
    plt.ylabel('Actual')
    plt.title(f'{name} Confusion Matrix')
    plt.show()

predict_and_plot(train_inputs, train_targets, name='Training')
```

```python
predict_and_plot(val_inputs, val_targets, name='Validation')
```

```python
predict_and_plot(test_inputs, test_targets, name='Test')
```

Compare to dumb models

```python
accuracy_score(test_targets, np.random.choice(["No","Yes"], len(test_inputs)))
```

```python
accuracy_score(test_targets, np.full(len(test_inputs), "No"))
```
